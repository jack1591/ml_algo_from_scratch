```python
import pandas as pd
import numpy as np
```

### Напише собственную реализацию KMEANS

```python
from sklearn.base import BaseEstimator, ClusterMixin
import copy
class My_KMeans(BaseEstimator, ClusterMixin):
    def __init__(self, n_clusters = 1, max_iter = 300, p = 2.0):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.p = p

    @staticmethod
    def minkowski_distance(first, second, p):
        distance = np.sum(abs(first - second) ** p)
        return distance ** (1/p)

    def choose_cluster(self, point):
        return np.argmin([self.minkowski_distance(point, center, self.p) for center in self.cluster_centers_])

    def update_centers(self, points, labels):
        for cluster_num in range(self.n_clusters):
            cluster_points = points[labels == cluster_num]
            if len(cluster_points) > 0:
                self.cluster_centers_[cluster_num] = np.mean(cluster_points, axis = 0)
        return self.cluster_centers_
        
    def fit(self, X):
        n_samples = X.shape[0]
        #self.cluster_centers_ = X[list[range(self.n_clusters)]]
        cluster_centers_idx = np.random.permutation(n_samples)[:self.n_clusters]
        self.cluster_centers_ = copy.deepcopy(X[cluster_centers_idx])

        for _ in range(self.max_iter):
            labels = np.array([self.choose_cluster(point) for point in X])
            self.cluster_centers_ = self.update_centers(X, labels)
            
        return self
    
    def predict(self, X):
        return [self.choose_cluster(point) for point in X]

    def transform(self, X):
        n_samples = X.shape[0]
        distances = np.zeros((n_samples, self.n_clusters))

        for i in range(n_samples):
            for j, center in enumerate(self.cluster_centers_):
                distances[i, j] = self.minkowski_distance(X[i], center, self.p)

        return distances
```

## Сравним качество работы

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

x_blobs, _ = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.6, random_state = 0)

plt.figure(figsize = (16, 10))
plt.scatter(x_blobs[:, 0], x_blobs[:, 1], s = 50)
plt.show()
```

### Готовая реализация sklearn

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4)
kmeans.fit(x_blobs)
y_kmeans = kmeans.predict(x_blobs)

plt.figure(figsize = (16, 10))
plt.scatter(x_blobs[:, 0], x_blobs[:, 1], c = y_kmeans, s = 50, cmap = 'viridis')
plt.show()
```

### Моя реализация

```python
my_kmeans = My_KMeans(n_clusters = 4)
my_kmeans.fit(x_blobs)
y_my_kmeans = my_kmeans.predict(x_blobs)

plt.figure(figsize = (16, 10))
plt.scatter(x_blobs[:, 0], x_blobs[:, 1], c = y_my_kmeans, s = 50, cmap = 'viridis')
plt.show()
```
